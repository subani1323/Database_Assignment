```python
import json

# Select / create database
db = client["mydatabase"]

# Function to insert JSON data into MongoDB
def insert_json(file_path, collection_name):
    try:
        with open(file_path, "r") as file:
            data = json.load(file)

        collection = db[collection_name]

        if isinstance(data, list):
            result = collection.insert_many(data)
            print(f"Inserted {len(result.inserted_ids)} documents into {collection_name}")
        else:
            result = collection.insert_one(data)
            print(f"Inserted 1 document into {collection_name}")

    except Exception as e:
        print(f"Error inserting {collection_name}: {e}")


# Insert your uploaded JSON files
insert_json("/content/app_events.json", "app_events")
insert_json("/content/complaints.json", "complaints")
insert_json("/content/customer.json", "customers")
insert_json("/content/data_dictionary.json", "data_dictionary")
insert_json("/content/deliveries.json", "deliveries")
insert_json("/content/drivers.json", "drivers")
insert_json("/content/hubs.json", "hubs")
insert_json("/content/incident.json", "incidents")
insert_json("/content/orders.json", "orders")
insert_json("/content/vehicle.json", "vehicles")
```
